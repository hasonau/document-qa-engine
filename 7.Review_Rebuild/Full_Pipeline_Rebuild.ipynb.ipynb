```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import pdfplumber as pdf
```

```python
filename = "Muhammad-pages.pdf"
wholeText=""
with pdf.open(f"../data/building/{filename}") as pdfFile:
    for page in pdfFile.pages:
        wholeText+=page.extract_text()
```

```python
words = wholeText.split(" ")
```

```python
i=0
overlap=100
chunkSize=300

chunks=[]
while  i < len(words):
    chunks.append(" ".join(words[i:i+chunkSize]))
    i+=(chunkSize-overlap)
print(chunks)
```

```python
from sentence_transformers import SentenceTransformer as ST
model = ST("all-MiniLM-L6-v2")
```

```python
embeddings = model.encode(chunks)
```

```python
embeddings
```

```python
import faiss 

index = faiss.IndexFlatL2(embeddings.shape[1])
index.add(embeddings)
```

```python
query = "What happened when Muhammad went to Ta'if?"
query_embeddings =  model.encode([query])
index.search(query_embeddings,3)
```
